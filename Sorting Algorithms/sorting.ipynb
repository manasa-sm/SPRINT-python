```python
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
result1=[]
result2=[]
result3=[]
lengths=[1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000]
for count in lengths:
    a=[]
    for i in range(0,count):
        a.append(random.randint(0,1000))
    res1 = %timeit -n1 -r1 -o np.sort(a, kind='quicksort')
    result1.append(res1.average)
    res2 = %timeit -n1 -r1 -o np.sort(a, kind='mergesort')
    result2.append(res2.average)
    res3 = %timeit -n1 -r1 -o np.sort(a, kind='heapsort')
    result3.append(res3.average)    
```

```python
print(result1) #quicksort timings
print(result2) #mergesort timings
print(result3) #heapsort timings
```

```python
plt.style.use('seaborn-whitegrid')
plt.title('Sorting Algorithms')
plt.xlabel('Test length')
plt.ylabel('Runtime (s)')
plt.plot(lengths, result1, 'r', label='quicksort')#quicksort
plt.plot(lengths, result2, 'b', label='mergesort')#mergesort
plt.plot(lengths, result3, 'g', label='heapsort')#heapsort
plt.legend()
plt.show()
```
